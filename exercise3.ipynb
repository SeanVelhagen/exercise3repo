```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
```

```python
img = mpimg.imread('/home/sean/Pictures/dimash face photo.jpg')
```

```python
grayImg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

```python
faceCasc = cv.CascadeClassifier('/home/sean/\
anaconda3/lib/python3.7/site-packages/cv2/data/haarcascade_frontalface_default.xml')
```

```python
eyeCasc = cv.CascadeClassifier('/home/sean/anaconda3/lib/python3.7/site-packages/cv2/data/haarcascade_eye.xml')
```

```python
faces = faceCasc.detectMultiScale(grayImg, 1.3, 5)
```

```python
for (x,y,fwidth,fheight) in faces:
    cv.rectangle(img, (x, y), (x + fwidth, y + fheight), (255, 0, 0), 2)
    roiGray = grayImg[y: y + fheight, x: x + fwidth]
    roiColor = img[y: y + fheight, x: x + fwidth]
    eyes = eyeCasc.detectMultiScale(roiGray)
    for (ex, ey, ewidth, eheight) in eyes:
        cv.rectangle(roiColor, (ex, ey), (ex + ewidth, ey + eheight), (0, 255, 0), 2)
```
